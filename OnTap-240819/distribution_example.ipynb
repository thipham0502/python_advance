```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
iris = sb.load_dataset('iris')
iris.head()
```

```python
# 1. Plotting univariate distributions
sb.distplot(iris['petal_length']) 
```

```python
sb.distplot(iris['petal_length'], kde=False)
```

```python
sb.distplot(iris['petal_length'], hist=False)
```

```python
sb.distplot(iris['petal_length'], rug=True, hist=False)
```

```python
# 2. Plotting bivariate distributions
sb.jointplot(data=iris, x='petal_length', y='petal_width')
```
