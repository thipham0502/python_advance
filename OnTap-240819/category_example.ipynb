```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
tips = sb.load_dataset('tips')
```

```python
# 1. Plotting univariate distributions
sb.stripplot(x=tips['total_bill'], jitter=True)
```

```python
sb.swarmplot(x=tips['total_bill'])
```

```python
sb.boxplot(x=tips['total_bill'])
```

```python
sb.boxplot(y=tips['total_bill'])
```

```python
# 2. Plotting bivariate distributions
sb.stripplot(x="day", y="total_bill", data=tips)
```

```python
sb.stripplot(x="day", y="total_bill", data=tips, hue='sex')
```

```python
sb.boxplot(x="day", y="total_bill", data=tips)
```

```python
sb.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sb.catplot(x="sex", y="total_bill", data=tips, kind='swarm')
```

```python
sb.catplot(x="sex", y="total_bill", data=tips, kind='box')
```

```python
sb.catplot(x="sex", y="total_bill", data=tips, kind='swarm', hue='smoker')
```

```python
sb.catplot(x="sex", y="total_bill", data=tips, kind='swarm', hue='smoker', col='time')
```

```python
titanic = sb.load_dataset('titanic')
```

```python
sb.countplot(x='class', data=titanic)
```

```python
sb.countplot(x='class', data=titanic, hue='who')
```

```python
sb.countplot(y='class', data=titanic, hue='who')
```

```python
sb.countplot(x='who', data=titanic)
```

```python
# height: tỉ lệ chiều cao các cột
# aspect: tỉ lệ chiều rộng các cột
sb.catplot(x="class", hue="who", col="survived", data=titanic, kind="count", height=4, aspect=.8)
```
