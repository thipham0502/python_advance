```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
tips = sb.load_dataset('tips')
```

```python
sb.regplot(x="total_bill", y="tip", data=tips)
```

```python
x_val = [95,85,80,70,60]
y_val = [85,95,70,65,70]
df = pd.DataFrame({'X':x_val, 'Y':y_val})
df
```

```python
sb.regplot(x="X", y="Y", data=df)
```

```python
sb.lmplot(x="total_bill", y="tip", data=tips)
```

```python
sb.lmplot(x="total_bill", y="tip", data=tips, hue='sex')
```
