```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Đọc file group_by_dep trong thư mục vi_du_06
df = pd.read_excel('vi_du_06/group_by_dep.xls')
df
```

```python
# Vẽ biểu đồ biểu diễn số nhân viên mỗi phòng
fix,ax = plt.subplots(ncols=2,figsize=(8,6))

## Ko chỉ định x
y = df['count_emp']
ax[0].plot(y,color='b') # ko chỉ định x thì lấy index của series y

## Có chỉ định x
y = df['count_emp']
x = np.arange(len(y))
ax[1].plot(x,y,color='g')
plt.show()
```

```python
# Vẽ biểu đồ thể hiện số lượng nhân viên theo phòng
x = range(len(y))
y = df['count_emp']
plt.figure(figsize=(14,5))
plt.plot(x,y,color='y')
plt.title('Thống kê số lượng nhân viên theo phòng',color='red',size=16)
plt.xlabel('Phòng',color='m',size=14)
plt.ylabel('Số nhân viên',color='m',size=14)
plt.xticks(x,df['DEPARTMENT_NAME'])
plt.grid(True)
plt.show()
```

```python
# Vẽ biểu đồ thể hiện tổng lương nhân viên theo phòng
x = range(len(y))
y = df['sum_sal']
plt.figure(figsize=(14,10))
plt.plot(x,y,color='c')
plt.title('Thống kê tổng lương nhân viên theo phòng',color='red',size=16)
plt.xlabel('Phòng',color='m',size=14)
plt.ylabel('Tổng lương',color='m',size=14)
plt.xticks(x,df['DEPARTMENT_NAME'])
plt.yticks(y, y.map(lambda x: '{:15,.0f}'.format(x))) # định dạng tiền tệ
plt.grid(True)
plt.show()
```

```python
# Vẽ biểu đồ kết hợp số lượng nhân viên và tổng lương theo phòng
plt.figure(figsize=(14,8))

x = range(len(y))
y = df['count_emp']*1000
y1 = df['sum_sal']
plt1 = plt.plot(x,y,color='g',label='Số nhân viên')
plt2 = plt.plot(x,y1,color='b',label='Tổng lương')

plt.title('Thống kê tổng lương và số lượng nhân viên nhân viên theo phòng',color='red',size=16)
plt.xlabel('Phòng',color='m',size=14)
plt.ylabel('Tổng lương & Số lượng nhân viên',color='m',size=14)
plt.xticks(x,df['DEPARTMENT_NAME'])

plt.legend(title='Ghi chú',loc=1)
plt.grid(True)
plt.show()
```

```python
# Vẽ lương thấp nhất, lương trung bình, lương lớn nhất của các phòng
## Cách thủ công
i = 7 # phòng Sales
y = df.iloc[i][['min_sal','avg_sal','max_sal']]
plt.plot(y, color='green', marker='o', markersize=10, label=df.iloc[i]['DEPARTMENT_NAME'])

j = 4 # phòng Shipping
y_ = df.iloc[j][['min_sal','avg_sal','max_sal']]

plt.plot(y_, color='blue', marker='o', markersize=10, label=df.iloc[j]['DEPARTMENT_NAME'])

plt.legend(title='Ghi chú')
plt.show()
```

```python
# Vẽ lương thấp nhất, lương trung bình, lương lớn nhất của các phòng
## Dùng vòng lặp
plt.figure(figsize=(12,10))
for i in range(df.shape[0]):
    y = df.iloc[i][['min_sal','avg_sal','max_sal']]
    plt.plot(y, marker='o', markersize=10, label=df.iloc[i]['DEPARTMENT_NAME'])

plt.xticks(['min_sal','avg_sal','max_sal'],['thấp nhất', 'trung bình', 'lớn nhất'])
plt.legend(title='Ghi chú')
plt.show()
```
