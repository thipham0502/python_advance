```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Line plot: pyplot.plot

### a. Dạng đường thẳng liền nét

```python
fig, ax = plt.subplots(figsize=(7,5)) # Thay đổi kích thước hình

year = np.array([2010,2011,2012,2013,2014])
populations = np.array([200,300,450,500,650])
plt.plot(year, populations, color = 'c', linewidth=1.5, label='Bieu_do_dan_so')

# Thay đổi chỉ số ở trục tọa độ
plt.xticks(year, color='m')
plt.yticks(np.arange(np.amin(populations),np.amax(populations),100), color='m')

# Đặt tên cho biểu đồ
plt.title('Biểu đồ dân số thê giới năm 2000 liền nét', color='g')

# Đặt tên cho trục tọa độ
plt.xlabel('Năm', color='b', size=12)
plt.ylabel('Dân số', color='b', size=12)

# Thêm ghi chú, hiển thị label của đường
plt.legend(title='Chu thich')

plt.show()
```

```python
# Vẽ nhiều đường trên 1 hình
fig, ax = plt.subplots(figsize=(8,6)) # Thay đổi kích thước hình

x = np.random.randint(50, size=10) # random 1 array 10 phần tử ngẫu nhiên từ 0-50
y1 = 3*x + 1
y2 = 5*x + 10

plt.plot(x, y1, color = 'b', linewidth=1, label='$y1 = 3*x + 1$')
plt.plot(x, y2, color = 'm', linewidth=1, label='$y2 = x - 9$')

# Thay đổi chỉ số ở trục tọa độ
plt.xticks(np.arange(np.amin(x), np.amax(x), 5))

# Đặt tên cho biểu đồ
plt.title('Biểu đồ 2 đường thẳng liền nét', color='g')

# Đặt tên cho trục tọa độ
plt.xlabel('x', size=14)
plt.ylabel('y', size=14)

# Thêm ghi chú, hiển thị label của đường
plt.legend(title='Chu thich')

plt.show()
```

### b. Dạng đường thẳng không liền nét

```python
fig, ax = plt.subplots(figsize=(8,6)) # Thay đổi kích thước hình

x = np.random.randint(50, size=10) # random 1 array 10 phần tử ngẫu nhiên từ 0-50
y1 = 3*x + 1
y2 = 5*x + 10

# Vẽ đường thẳng ko liền nét
plt.plot(x, y1, 'b*', label='$y1 = 3*x + 1$') # 'bo': ký hiệu điểm là '*' màu xanh dương (Blue)
plt.plot(x, y2, 'cx', label='$y2 = x - 9$') # 'cx': ký hiệu điểm là 'x' màu xanh ngọc (Cyan)

# Thay đổi chỉ số ở trục tọa độ
plt.xticks(np.arange(np.amin(x), np.amax(x), 5))

# Đặt tên cho biểu đồ
plt.title('Biểu đồ 2 đường thẳng không liền nét', color='g')

# Đặt tên cho trục tọa độ
plt.xlabel('x', size=14)
plt.ylabel('y', size=14)

# Thêm ghi chú, hiển thị label của đường
plt.legend(title='Chu thich')

plt.show()
```

### c. Dạng đường thẳng liền nét và có thể hiện điểm dữ liệu

```python
fig, ax = plt.subplots(figsize=(8,6)) # Thay đổi kích thước hình

x = np.array([1,2,3,4,5])
y1 = 3*x + 1
y2 = 5*x + 3

# Vẽ
# linestyles = ['-', '--', '-.', ':']
plt.plot(x, y1, color='m', marker='o', linestyle='-.', linewidth=1, markersize=8, label='$y1 = 3*x + 1$')
plt.plot(x, y2, color='orange', marker='s', linestyle='--', linewidth=1, markersize=8, label='$y2 = 5*x + 3$')

# Thay đổi chỉ số ở trục tọa độ
plt.xticks(np.arange(np.amin(x), np.amax(x), 5))

# Đặt tên cho biểu đồ
plt.title('Biểu đồ 2 đường thẳng liền nét và có thể hiện điểm dữ liệu', color='g', size=14)

# Đặt tên cho trục tọa độ
plt.xlabel('x', color='b', size=14)
plt.ylabel('y', color='b', size=14)

# Thêm ghi chú, hiển thị label của đường
plt.legend(title='Chu thich')

plt.show()
```

## Area plot: pyplot.fill_between
Biểu đồ tương tự như line plot nhưng vùng giữa trục x và line được tô màu

```python
fig, ax = plt.subplots(figsize=(8,6)) # Thay đổi kích thước hình

x = np.array([1,2,3,4,5])
y1 = np.array([223,142,647,241,533])

# Vẽ area plot
plt.fill_between(x, y1, color='skyblue', alpha=0.5) # alpha: độ đậm nhạt
# Thêm đường thẳng viền ngoài
plt.plot(x, y1, color='b', linewidth=1.5, label='$y1 = 3*x + 1$') # alpha: độ đậm nhạt
# Thêm lưới
plt.grid(True)

# Thay đổi chỉ số ở trục tọa độ
plt.xticks(x)

# Đặt tên cho biểu đồ
plt.title('Biểu đồ area', color='g', size=14)

# Đặt tên cho trục tọa độ
plt.xlabel('x', color='m', size=14)
plt.ylabel('y', color='m', size=14)

# Thêm ghi chú, hiển thị label của đường
plt.legend(title='Chu thich')

plt.show()
```
