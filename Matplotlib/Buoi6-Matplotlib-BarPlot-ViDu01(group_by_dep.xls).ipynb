```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('group_by_dep.xls')
print(df)
```

```python
y = df['sum_sal']
x = range(len(y))
plt.figure(figsize=(16,5))
plt.bar(x,y)
plt.title('Thống kê tổng lương theo phòng', color='r')
plt.xlabel('Phòng', color='r')
plt.ylabel('Tổng lương', color='r')
plt.xticks(x, df['DEPARTMENT_NAME'])
plt.show()
```

```python
# Vẽ biểu đồ thể hiện tổng lương nhân viên theo phòng
x = df['DEPARTMENT_NAME']
y = df['sum_sal']
plt.figure(figsize=(16,5))
plt.bar(x,y,color='m')
plt.title('Thống kê tổng lương theo phòng', color='r')
plt.xlabel('Phòng', color='r')
plt.ylabel('Tổng lương', color='r')
plt.show()
```
