```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Đọc số liệu trong tập tin years.txt trong thư mục vi_du_06
f = open('vi_du_06/years.txt')
data = f.read()
f.close()

years = [int(y) for y in data.split(',')]
years = pd.Series(years)
print(years)
```

```python
# Đọc số liệu trong tập tin pops.txt trong thư mục vi_du_06
f = open('vi_du_06/pops.txt')
data = f.read()
f.close()

pops = [float(p) for p in data.split(',')]
pops = pd.Series(pops)
print(pops)
```

```python
# Plot
plt.plot(years, pops)
plt.show()
```

```python
# Vẽ số liệu từ năm 1950-1954
# *Cách 1:
### Lấy các năm từ 1950-1954
indexes = np.where((years>=1950) & (years<=1954))[0]
x = years[indexes]
# print(x)

### Lấy dân số từ 1950-1954
y = pops[indexes]
# print(y)

### Vẽ
plt.plot(x,y,color='orange')
```

```python
# Vẽ số liệu từ năm 1950-1954
# *Cách 2:
### Lấy các năm từ 1950-1954
dieu_kien = (years>=1950) & (years<=1954)
x = years[dieu_kien]
# print(x)

### Lấy dân số từ 1950-1954
y = pops[dieu_kien]
# print(y)

### Vẽ
plt.plot(x,y,color='b')
```

```python
### Vẽ
plt.plot(x,y,color='g',marker='o',markersize=10,linestyle='dashed')
```

```python
# Vẽ số liệu từ năm 2005-2019
### Lấy các năm từ 2005-2019
dieu_kien = (years>=2005) & (years<=2019)
x = years[dieu_kien]
# print(x)

### Lấy dân số từ 1950-1954
y = pops[dieu_kien]
# print(y)

### Vẽ
plt.xticks(np.arange(np.amin(x), np.amax(x) + 1, 2))
plt.plot(x,y,color='m',marker='*',markersize=10,linestyle=':')
```

```python
# Hàm lấy số liệu theo năm
def get_data(from_year, to_year):
    dieu_kien = (years>=from_year) & (years<=to_year)
    x = years[dieu_kien]
    y = pops[dieu_kien]
    return (x,y)
```

```python
# Hàm vẽ theo bộ dữ liệu
def plot_data(x,y):
    plt.plot(x,y,color='c',marker='h',markersize=10,linestyle='-.')
    plt.show()
```

```python
# Vẽ
x,y = get_data(1965,1970)
plot_data(x,y)
```
