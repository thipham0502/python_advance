```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Line plot

```python
year = np.array([2010,2011,2012,2013,2014])
populations = np.array([200,300,450,500,650])
plt.plot(year, populations, color = 'c', linewidth=1.5, label='y')

# Thay đổi chỉ số ở trục tọa độ
plt.xticks(year)
plt.yticks(np.arange(np.amin(populations),np.amax(populations),100))

# Đặt tên cho trục tọa độ
plt.xlabel('Nam')
plt.ylabel('Dan so')

# Thêm ghi chú, hiển thị label của đường
plt.legend(title='Chu thich')

#
plt.show()
```

### Bar plot

```python
plt.bar(year,populations)
plt.show()
```

```python
# Đọc file excel employees.xls trong thư mục thuc_hanh_1
emps = pd.read_excel('thuc_hanh_1/employees.xls')

# Đọc file excel departments.xls trong thư mục thuc_hanh_1
deps = pd.read_excel('thuc_hanh_1/departments.xls')

### Chuẩn bị data
# Kết 2 bảng nhân viên. phòng ban theo mã phòng ban
# lấy hết các cột trong bảng nhân viên và chỉ lấy 2 cột là mã phòng và tên phòng trong bảng phòng ban
df_NV_PB = pd.merge(emps, deps[['DEPARTMENT_ID','DEPARTMENT_NAME']], on=['DEPARTMENT_ID'], how='left')

# Tính tổng lương và số nhân viên
df_sum = df_NV_PB.groupby('DEPARTMENT_ID', as_index=False)['SALARY'].sum()
# print(df_sum)
df_count = df_NV_PB.groupby('DEPARTMENT_ID', as_index=False)['SALARY'].count()
# print(df_count)

fig, ax = plt.subplots(figsize=(15,7))
bar1 = plt.bar(df_sum['DEPARTMENT_ID'], df_sum['SALARY'], color='c', width=3, edgecolor='k', align='center')
bar2 = plt.bar(df_count['DEPARTMENT_ID'], df_count['SALARY']*1000, color='y', width=3, edgecolor='k', align='center')
plt.legend([bar1,bar2],['sum','count'],loc=1,title='Chu thich')
plt.show()
```
