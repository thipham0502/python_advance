```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = open('years.txt')
data = f.read()
f.close()
```

```python
years = [int(y) for y in data.split(',')]
years = pd.Series(years)
```

```python
f = open('pops.txt')
data = f.read()
f.close()
```

```python
pops = [float(p) for p in data.split(',')]
pops = pd.Series(pops)
```

```python
year = years[(years>=2010) & (years<=2016)]
pop = pops[0:len(year)]
plt.bar(year, pop)
plt.title('Biểu đồ dân số thế giới', color='red')
plt.xlabel('Năm', color='red')
plt.ylabel('Dân số', color='red')
plt.show()
```

```python
df = pd.read_excel('group_by_dep.xls')
print(df)
```

```python
y = df['sum_sal']
x = range(len(y))
plt.figure(figsize=(16,5))
plt.bar(x,y)
plt.title('Thống kê tổng lương theo phòng', color='red')
plt.xlabel('Phòng', color='red')
plt.ylabel('Tổng lương', color='red')
plt.xticks(x, df['DEPARTMENT_NAME'])
plt.show()
```

```python
x = df['DEPARTMENT_NAME']
y = df['sum_sal']
plt.figure(figsize=(16,5))
plt.bar(x,y,color='m')
plt.show()
```

```python
x = [1]
y = [df.iloc[4]['sum_sal']]
bar_1 = plt.bar(x,y,color='r')
x_ = [2]
y_ = [df.iloc[4]['count_emp']*1000]
bar_2 = plt.bar(x_,y_,color='g')
plt.xticks(x, [df.iloc[4]['DEPARTMENT_NAME'],df.iloc[7]['DEPARTMENT_NAME']])
plt.legend([bar_1,bar_2],['Tổng lương','Số nhân viên'],loc=1)
plt.show()
```

```python
x = [1,3]
y = [df.iloc[4]['sum_sal'],df.iloc[7]['sum_sal']]
bar_1 = plt.bar(x,y,color='r')
x_ = [2,4]
y_ = [df.iloc[4]['count_emp']*1000,df.iloc[7]['count_emp']*1000]
bar_2 = plt.bar(x_,y_,color='g')
plt.xticks(x, [df.iloc[4]['DEPARTMENT_NAME'],df.iloc[7]['DEPARTMENT_NAME']])
plt.legend([bar_1,bar_2],['Tổng lương','Số nhân viên'],loc=1)
plt.show()
```

```python
x = np.arange(1,df.shape[0]*2,2)
y = df['sum_sal']
plt.figure(figsize=(16,5))
bar_1 = plt.bar(x,y,color='r')
x_ = np.arange(2,df.shape[0]*2+1,2)
y_ = df['count_emp']*1000
bar_2 = plt.bar(x_,y_,color='g')
plt.xticks(x, df['DEPARTMENT_NAME'])
plt.legend([bar_1,bar_2],['Tổng lương','Số nhân viên'],loc=1)
plt.show()
```

```python
dep_names = df['DEPARTMENT_NAME']
emp_count = df['count_emp']
y_pos = np.arange(len(dep_names))
plt.barh(y_pos,emp_count,color='green',ecolor='black')
plt.yticks(y_pos,dep_names)
plt.xlabel('Số nhân viên',color='red')
plt.title('Đếm nhân viên theo phòng',color='blue')
plt.show()
```
