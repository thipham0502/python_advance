```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Biểu đồ phân tán: Scatter plot

```python
x = np.random.randint(50, size=10) # random 10 giá trị trong khoảng 0-49
y = np.random.randint(50, size=10) # random 10 giá trị trong khoảng 0-49

fig, ax = plt.subplots(figsize=(8,6))

plt.scatter(x,y)
plt.show()
```

```python
x = np.random.randint(50, size=10) # random 10 giá trị trong khoảng 0-49
y = np.random.randint(50, size=10) # random 10 giá trị trong khoảng 0-49

fig, ax = plt.subplots(figsize=(8,6))

plt.scatter(x, y, color='b', alpha=0.5, marker=r'$\clubsuit$', label='Value')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(title='Ghi chú', loc=4)
plt.show()
```

### Dạng bubble chart

```python
colors = np.random.rand(10)
area = (30 * np.random.rand(10))**2

plt.scatter(x, y, s=area, c=colors, alpha=0.7)
plt.show()
```
