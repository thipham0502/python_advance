```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

### Câu 1: vẽ barplot cho dữ liệu x,y

```python
x = ['Excellent','Very Good','Good','Medium']
y = [5.0,15.0,20.0,5.0]

# Cách 1: ko tạo dataframe
sb.barplot(x,y)
```

```python
# Cách 2: tạo dataframe
df = pd.DataFrame({'type':x, 'amount':y}, columns=['type','amount'])
sb.barplot(x='type',y='amount',data=df)
```

### Câu 2: vẽ barplot & swarmplot cho dữ liệu chars

```python
chars = 'A A A A A A A A A A B B B B B B C C C C C C '\
'D D D D D D D D D D E E E E E E E F F F F '\
'G G G G G G G H H H H'
print(chars)
```

```python
# Vẽ barplor với mỗi cột có tên là 1 ký tự,
# giá trị là số lần xuất hiện của ký tự đó trong chuỗi
list_chars = chars.split()
# print(list_chars)
list_count = np.ones(len(list_chars))

df_chars = pd.DataFrame({'index':list_chars,'count':list_count},columns=['index','count'])
# print(df_chars)
df_count = df_chars.groupby('index').count()
df_count['char']=df_count.index # tạo cột mới có giá trị giống với index để vẽ
print(df_count)
```

```python
sb.set()
sb.barplot(x='char',y='count',data=df_count)
plt.show()
```

```python
sb.swarmplot(x='index',y='count',data=df_chars)
```
