```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# Load database "iris"
iris = sb.load_dataset('iris')
print(iris.shape)
print(iris.head())
print(iris.tail())
```

```python
# Load database "tips"
tips = sb.load_dataset('tips')
print(tips.shape)
print(tips.head())
print(tips.tail())
```

```python
# Load database "titanic"
titanic = sb.load_dataset('titanic')
print(titanic.shape)
print(titanic.head())
print(titanic.tail())
```

```python
# displot
plt.figure(figsize=(10,8))

# kde
sb.distplot(tips['total_bill'])
```

```python
# displot
plt.figure(figsize=(10,8))

# no kde
sb.distplot(tips['total_bill'], kde=False)
plt.show()
```

```python
# Phân phối chuẩn
heights = [74,74,72,72,73,69,69,71,76,71,73,73,74,76,73,72,71,70]
plt.figure(figsize=(10,8))
sb.distplot(heights)
```

```python
# Rug
heights = [74,74,72,72,73,69,69,71,76,71,73,73,74,76,73,72,71,70,73,72,69,70,74,72]
plt.figure(figsize=(10,8))
sb.distplot(heights, hist=False, rug=True)
```

```python
# Rug
plt.figure(figsize=(10,8))
sb.distplot(tips['total_bill'], hist=False, rug=True)

print(tips['total_bill'].describe())
```

```python
# Đếm số lượng tiền hóa đơn > 14 và < 20
df_tips_total_bill = tips[(tips['total_bill'] > 14) & (tips['total_bill'] < 20)]
df_tips_total_bill.groupby('total_bill').tip.count()
```

```python
# Vẽ đường thằng hồi quy cho data tips
sb.regplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
# regplot: Vẽ đường thằng hồi quy cho data iris
sb.regplot(data = iris, x = 'petal_length', y = 'petal_width')
```

```python
# lmplot (nâng cao của regplot)
# Vẽ đường thằng hồi quy cho data iris
## hue: pha màu cho từng đối tượng trong 1 nhóm
sb.lmplot(data=iris, x='petal_length', y='petal_width', hue='species')
```

```python
# lmplot (nâng cao của regplot)
# Vẽ đường thằng hồi quy cho data iris
## col/row: vẽ phân biệt các đối tượng trong nhóm trên 1 dòng/cột
plt.figure(figsize=(10,8))
sb.lmplot(data=iris, x='petal_length', y='petal_width', col='species')
```

```python
# lmplot (nâng cao của regplot)
# Vẽ đường thằng hồi quy cho data iris
## col/row: vẽ phân biệt các đối tượng trong nhóm trên 1 dòng/cột
plt.figure(figsize=(10,8))
sb.lmplot(data=iris, x='petal_length', y='petal_width', row='species')
```

```python
# lmplot (nâng cao của regplot)
# Vẽ đường thằng hồi quy cho data tips
## hue: pha màu cho từng đối tượng trong 1 nhóm
sb.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
# Đếm số lượng nam và nữ cho tip
tips.groupby('sex').tip.count()
```

```python
# Tồng số tiền tip nam và nữ cho
tips.groupby('sex').total_bill.sum()
```

```python
# Scatter tổng tiền hóa đơn trong ngày
plt.figure(figsize=(10,8))
sb.swarmplot(data=tips, x='day', y='total_bill')
```

```python
# Scatter tổng tiền tip trong buổi
plt.figure(figsize=(10,8))
sb.swarmplot(data=tips, x='time', y='tip')
```

```python
# Scatter độ tuổi của các khoang tàu
plt.figure(figsize=(10,8))
sb.swarmplot(data=titanic, x='class', y='age')
```

```python
# Scatter độ tuổi của các khoang tàu, phân loại theo giới tính
plt.figure(figsize=(10,8))
sb.swarmplot(data=titanic, x='class', y='age', hue='sex')
```

```python
# Barplot độ tuổi của những người sống sót/ko sống sót
plt.figure(figsize=(10,8))
sb.barplot(data=titanic, x='survived', y='age')
```

```python
# Đếm số người ko được cứu theo giới tính
titanic[titanic['survived']==0].groupby('sex').count()
```

```python
# Đếm số người được cứu theo giới tính
titanic[titanic['survived']==1].groupby('sex').count()
```

```python
# Barplot tỉ lệ % người sống sót theo giới tính
# Male: 109/(109+468)*100 = 18.89%
# Female: 233/(81+233)*100 = 74.2%
plt.figure(figsize=(10,8))
## xóa vạch đen trên đỉnh cột => ci = None
# sb.barplot(data=titanic, x='sex', y='survived', ci=None)
sb.barplot(data=titanic, x='survived', y='sex', ci=None)
```

## Seaborn styles

```python
# Scatter độ tuổi của các khoang tàu
# Định dạng màu đậm dần theo khối lượng dữ liệu
plt.figure(figsize=(10,8))

## Set color palette
sb.set_palette('Reds')

sb.swarmplot(data=titanic, x='class', y='age', hue='sex')
```

```python
# Barplot tỉ lệ % người sống sót theo giới tính
# Định dạng màu đậm dần theo khối lượng dữ liệu
plt.figure(figsize=(10,8))

## Set color palette
sb.set_palette('Blues')
sb.barplot(data=titanic, x='survived', y='sex', ci=None)
```

```python
# Scatter độ tuổi của các khoang tàu & chia màu theo giới tính
# Định dạng màu đậm dần theo khối lượng dữ liệu
plt.figure(figsize=(10,8))

## Set chế độ seaborn styles
sb.set()

sb.swarmplot(data=titanic, x='class', y='age', hue='sex')
```

```python
# Scatter tổng tiền hóa đơn trong buổi & chia màu theo giới tính
# Định dạng màu đậm dần theo khối lượng dữ liệu
plt.figure(figsize=(10,8))

## Set chế độ seaborn styles
sb.set()

sb.swarmplot(data=tips, x='time', y='total_bill', hue='sex')
```

```python
# Box plot
plt.figure(figsize=(10,8))
sb.boxplot(data=iris,x='species',y='petal_length')
```

```python
iris.groupby('species')['petal_length'].describe()
```

```python
# Violin plot (lồng boxplot ở trong, chấm trắng ở giữa là mean)
plt.figure(figsize=(10,8))
sb.violinplot(data=iris, x='petal_length', y='species')
```

```python
# Boxenplot
plt.figure(figsize=(10,8))
sb.boxenplot(data=iris, x='petal_length', y='species')
```

```python
# Boxenplot
plt.figure(figsize=(12,8))
sb.boxenplot(data=tips, x='tip', y='sex', hue='samoker')
```

```python
# Pointplot (đoạn thẳng là khoảng tin cậy, chấm tròn là mean)
plt.figure(figsize=(12,8))
sb.pointplot(data=iris, x='species', y='petal_length', color='r', capsize = 0.5)
```
