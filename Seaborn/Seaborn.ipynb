```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# Load database "iris"
iris = sb.load_dataset('iris')
print(iris.shape)
print(iris.head())
print(iris.tail())
```

```python
# Load database "tips"
tips = sb.load_dataset('tips')
print(tips.shape)
print(tips.head())
print(tips.tail())
```

```python
# Load database "titanic"
titanic = sb.load_dataset('titanic')
print(titanic.shape)
print(titanic.head())
print(titanic.tail())
```

```python
# displot
plt.figure(figsize=(10,8))

# kde
sb.distplot(tips['total_bill'])
```

```python
# displot
plt.figure(figsize=(10,8))

# no kde
sb.distplot(tips['total_bill'], kde=False)
plt.show()
```

```python
# Phn phi chun
heights = [74,74,72,72,73,69,69,71,76,71,73,73,74,76,73,72,71,70]
plt.figure(figsize=(10,8))
sb.distplot(heights)
```

```python
# Rug
heights = [74,74,72,72,73,69,69,71,76,71,73,73,74,76,73,72,71,70,73,72,69,70,74,72]
plt.figure(figsize=(10,8))
sb.distplot(heights, hist=False, rug=True)
```

```python
# Rug
plt.figure(figsize=(10,8))
sb.distplot(tips['total_bill'], hist=False, rug=True)

print(tips['total_bill'].describe())
```

```python
# m s lng tin ha n > 14 v < 20
df_tips_total_bill = tips[(tips['total_bill'] > 14) & (tips['total_bill'] < 20)]
df_tips_total_bill.groupby('total_bill').tip.count()
```

```python
# V ng thng hi quy cho data tips
sb.regplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
# regplot: V ng thng hi quy cho data iris
sb.regplot(data = iris, x = 'petal_length', y = 'petal_width')
```

```python
# lmplot (nng cao ca regplot)
# V ng thng hi quy cho data iris
## hue: pha mu cho tng i tng trong 1 nhm
sb.lmplot(data=iris, x='petal_length', y='petal_width', hue='species')
```

```python
# lmplot (nng cao ca regplot)
# V ng thng hi quy cho data iris
## col/row: v phn bit cc i tng trong nhm trn 1 dng/ct
plt.figure(figsize=(10,8))
sb.lmplot(data=iris, x='petal_length', y='petal_width', col='species')
```

```python
# lmplot (nng cao ca regplot)
# V ng thng hi quy cho data iris
## col/row: v phn bit cc i tng trong nhm trn 1 dng/ct
plt.figure(figsize=(10,8))
sb.lmplot(data=iris, x='petal_length', y='petal_width', row='species')
```

```python
# lmplot (nng cao ca regplot)
# V ng thng hi quy cho data tips
## hue: pha mu cho tng i tng trong 1 nhm
sb.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
# m s lng nam v n cho tip
tips.groupby('sex').tip.count()
```

```python
# Tng s tin tip nam v n cho
tips.groupby('sex').total_bill.sum()
```

```python
# Scatter tng tin ha n trong ngy
plt.figure(figsize=(10,8))
sb.swarmplot(data=tips, x='day', y='total_bill')
```

```python
# Scatter tng tin tip trong bui
plt.figure(figsize=(10,8))
sb.swarmplot(data=tips, x='time', y='tip')
```

```python
# Scatter  tui ca cc khoang tu
plt.figure(figsize=(10,8))
sb.swarmplot(data=titanic, x='class', y='age')
```

```python
# Scatter  tui ca cc khoang tu, phn loi theo gii tnh
plt.figure(figsize=(10,8))
sb.swarmplot(data=titanic, x='class', y='age', hue='sex')
```

```python
# Barplot  tui ca nhng ngi sng st/ko sng st
plt.figure(figsize=(10,8))
sb.barplot(data=titanic, x='survived', y='age')
```

```python
# m s ngi ko c cu theo gii tnh
titanic[titanic['survived']==0].groupby('sex').count()
```

```python
# m s ngi c cu theo gii tnh
titanic[titanic['survived']==1].groupby('sex').count()
```

```python
# Barplot t l % ngi sng st theo gii tnh
# Male: 109/(109+468)*100 = 18.89%
# Female: 233/(81+233)*100 = 74.2%
plt.figure(figsize=(10,8))
## xa vch en trn nh ct => ci = None
# sb.barplot(data=titanic, x='sex', y='survived', ci=None)
sb.barplot(data=titanic, x='survived', y='sex', ci=None)
```

## Seaborn styles

```python
# Scatter  tui ca cc khoang tu
# nh dng mu m dn theo khi lng d liu
plt.figure(figsize=(10,8))

## Set color palette
sb.set_palette('Reds')

sb.swarmplot(data=titanic, x='class', y='age', hue='sex')
```

```python
# Barplot t l % ngi sng st theo gii tnh
# nh dng mu m dn theo khi lng d liu
plt.figure(figsize=(10,8))

## Set color palette
sb.set_palette('Blues')
sb.barplot(data=titanic, x='survived', y='sex', ci=None)
```

```python
# Scatter  tui ca cc khoang tu & chia mu theo gii tnh
# nh dng mu m dn theo khi lng d liu
plt.figure(figsize=(10,8))

## Set ch  seaborn styles
sb.set()

sb.swarmplot(data=titanic, x='class', y='age', hue='sex')
```

```python
# Scatter tng tin ha n trong bui & chia mu theo gii tnh
# nh dng mu m dn theo khi lng d liu
plt.figure(figsize=(10,8))

## Set ch  seaborn styles
sb.set()

sb.swarmplot(data=tips, x='time', y='total_bill', hue='sex')
```
