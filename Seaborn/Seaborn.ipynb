```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# Load database "titanic"
titanic = sb.load_dataset('titanic')
print(titanic.shape)
print(titanic.head())
print(titanic.tail())
```

## Data Aware Grid

### Facetgrid

```python
# V biu  phn tn biu din  di cnh hoa 
# cho tng loi hoa trong data iris bng FacetGrid
plt.figure(figsize=(12,10))

## col: trc nm ngang
# To grid gm 3 biu  tng ng vi mi loi hoa trong data iris
my_grid = sb.FacetGrid(data=iris, col='species')

# Xc nh loi biu  cho tng grid con v v theo petal_length ( y chn biu  phn tn swarmplot)
my_grid.map(sb.swarmplot, 'petal_length')
```

```python
# V biu  phn tn biu din  di cnh hoa 
# cho tng loi hoa trong data iris bng FacetGrid
plt.figure(figsize=(12,10))

## row: trc nm dc
# To grid gm 3 biu  tng ng vi mi loi hoa trong data iris
my_grid = sb.FacetGrid(data=iris, row='species')

# Xc nh loi biu  cho tng grid con v v theo petal_length ( y chn biu  phn tn swarmplot)
my_grid.map(sb.swarmplot, 'petal_length')
```

```python
# V biu  boxplot biu din tng tin ha n
# cho tng bui (time) trong tng ngy (day) 
# trong data tips bng FacetGrid
plt.figure(figsize=(12,10))

## col: nm ngang
# To grid
my_grid_1 = sb.FacetGrid(data=tips, col='time', row='day')

# Xc nh loi biu  cho tng grid con v v theo total_bill 
# ( y chn biu  boxplot)
my_grid_1.map(sb.boxplot, 'total_bill')
```

```python
# V biu  scatter biu din 
# s tin tip theo tng tin ha n
# cho tng bui (time) trong data tips bng FacetGrid
plt.figure(figsize=(12,10))

# To grid gm cc bui (time)
my_grid_2 = sb.FacetGrid(data=tips, col='time')

# Xc nh loi biu  cho tng grid con 
# v s tin tip (y) theo tng tin ha n (x)
# Chia mu theo gii tnh (hue)
# ( y chn biu  scatterplot)
my_grid_2.map(sb.scatterplot, 'total_bill', 'tip', 'sex')
```
