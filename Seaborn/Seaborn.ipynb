```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# Load database "iris"
iris = sb.load_dataset('iris')
print(iris.shape)
print(iris.head())
print(iris.tail())
```

```python
# Load database "tips"
tips = sb.load_dataset('tips')
print(tips.shape)
print(tips.head())
print(tips.tail())
```

```python
# Load database "titanic"
titanic = sb.load_dataset('titanic')
print(titanic.shape)
print(titanic.head())
print(titanic.tail())
```

```python
# displot
plt.figure(figsize=(10,8))

# kde
sb.distplot(tips['total_bill'])
```

```python
# Vẽ đường thằng hồi quy cho data tips
sb.regplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
# regplot: Vẽ đường thằng hồi quy cho data iris
sb.regplot(data = iris, x = 'petal_length', y = 'petal_width')
```

```python
# lmplot (nâng cao của regplot)
# Vẽ đường thằng hồi quy cho data iris
## hue: pha màu cho từng đối tượng trong 1 nhóm
sb.lmplot(data=iris, x='petal_length', y='petal_width', hue='species')
```

```python
# lmplot (nâng cao của regplot)
# Vẽ đường thằng hồi quy cho data iris
## col/row: vẽ phân biệt các đối tượng trong nhóm trên 1 dòng/cột
plt.figure(figsize=(10,8))
sb.lmplot(data=iris, x='petal_length', y='petal_width', row='species')
```

## Data Aware Grid

### Facetgrid

```python
# Vẽ biểu đồ phân tán biểu diễn độ dài cánh hoa 
# cho từng loại hoa trong data iris bằng FacetGrid
plt.figure(figsize=(12,10))

## col: trục nằm ngang
# Tạo grid gồm 3 biểu đồ tương ứng với mỗi loại hoa trong data iris
my_grid = sb.FacetGrid(data=iris, col='species')

# Xác định loại biểu đồ cho từng grid con và vẽ theo petal_length (ở đây chọn biểu đồ phân tán swarmplot)
my_grid.map(sb.swarmplot, 'petal_length')
```

```python
# Vẽ biểu đồ phân tán biểu diễn độ dài cánh hoa 
# cho từng loại hoa trong data iris bằng FacetGrid
plt.figure(figsize=(12,10))

## row: trục nằm dọc
# Tạo grid gồm 3 biểu đồ tương ứng với mỗi loại hoa trong data iris
my_grid = sb.FacetGrid(data=iris, row='species')

# Xác định loại biểu đồ cho từng grid con và vẽ theo petal_length (ở đây chọn biểu đồ phân tán swarmplot)
my_grid.map(sb.swarmplot, 'petal_length')
```

```python
# Vẽ biểu đồ boxplot biểu diễn tổng tiền hóa đơn
# cho từng buổi (time) trong từng ngày (day) 
# trong data tips bằng FacetGrid
plt.figure(figsize=(12,10))

## col: nằm ngang
# Tạo grid
my_grid_1 = sb.FacetGrid(data=tips, col='time', row='day')

# Xác định loại biểu đồ cho từng grid con và vẽ theo total_bill 
# (ở đây chọn biểu đồ boxplot)
my_grid_1.map(sb.boxplot, 'total_bill')
```

```python
# Vẽ biểu đồ scatter biểu diễn 
# số tiền tip theo tổng tiền hóa đơn
# cho từng buổi (time) trong data tips bằng FacetGrid
plt.figure(figsize=(12,10))

# Tạo grid gồm các buổi (time)
my_grid_2 = sb.FacetGrid(data=tips, col='time')

# Xác định loại biểu đồ cho từng grid con 
# vẽ số tiền tip (y) theo tổng tiền hóa đơn (x)
# Chia màu theo giới tính (hue)
# (ở đây chọn biểu đồ scatterplot)
my_grid_2.map(sb.scatterplot, 'total_bill', 'tip', 'sex')
```
