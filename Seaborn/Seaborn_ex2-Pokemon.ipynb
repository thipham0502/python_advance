```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

## Đọc file Pokemon.csv

```python
df_pokemon = pd.read_csv('chapter7/dataset/pokemon.csv',index_col=0) # Cột đầu tiên trong dữ liệu là index của df
df_pokemon
```

### a. Biểu đồ phân tán: phân loại với các điểm ko chồng chéo (=> dùng swarmplot), so sánh attack (trục x) và defense (trục y)

```python
plt.figure(figsize=(12,8))
sb.set()
sb.swarmplot(x='Attack',y='Defense',data=df_pokemon)
plt.show()
```

### b. Biểu đồ phân tán: so sánh attack (trục x) và defense (trục y), vẽ thêm 1 đường thẳng hồi quy lmplot

```python
plt.figure(figsize=(12,8))
sb.set()
sb.lmplot(x='Attack',y='Defense',data=df_pokemon)
plt.show()
```

### c. Vẽ đường thẳng hồi quy biểu diễn mối quan hệ tuyến tính giữa Attack(x) và Speed(y)

```python
plt.figure(figsize=(12,8))
sb.set()
sb.regplot(x='Attack',y='Speed',data=df_pokemon,color='m',marker='*')
plt.show()
```

### d. Vẽ boxplot cho toàn bộ dữ liệu

```python
plt.figure(figsize=(12,8))
sb.set()
sb.boxplot(data=df_pokemon)
plt.show()
```
